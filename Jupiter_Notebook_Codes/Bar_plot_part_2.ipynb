```python
# To make horizontal plot, just assign x with a numeric variable rather than categoric as well as y with categoric variable rather 
# than numeric variable 
import seaborn as sns
import matplotlib.pyplot as plt

ship = sns.load_dataset("titanic")
ship
sns.barplot(x = 'age', y = 'class', data=ship, saturation = 1)
plt.show()
```

```python
# Now dealing with edgecolor, errcolor, farecolor and linewidth...
# To make horizontal plot, just assign x with a numeric variable rather than categoric as well as y with categoric variable rather 
# than numeric variable 
import seaborn as sns
import matplotlib.pyplot as plt

ship = sns.load_dataset("titanic")
ship
sns.barplot(x = 'class', y = 'age', data=ship,
           linewidth = 5, facecolor = (1,1,1,1), 
            errcolor = '0', edgecolor = '0')
plt.show()
#changing the values within facecolor will change the color inside the bars' face...
```
