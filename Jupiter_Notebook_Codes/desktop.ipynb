## Giving limits 


```python
import seaborn as sns
import matplotlib.pyplot as plt
phool = sns.load_dataset("iris")
phool
sns.lineplot(x = 'petal_length', y ='petal_width', data= phool)
plt.title('Iris dataset plot')
plt.xlim(2)
plt.ylim(1)
plt.show()
```

## size setting

```python
import seaborn as sns
import matplotlib.pyplot as plt
phool = sns.load_dataset("iris")
phool
sns.lineplot(x = 'petal_length', y ='petal_width', data= phool)
plt.title('Iris dataset plot')
plt.xlim(2)
plt.ylim(1)
plt.figure(figsize=(7,9))
plt.show()
```
