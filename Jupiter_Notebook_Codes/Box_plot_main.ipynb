```python
# using hue+dodge+palette+color(through hex color picker(copypaste method))
# import libraries 
import seaborn as sns
sns.set(style='whitegrid')
import matplotlib.pyplot as plt


tip = sns.load_dataset("tips")

sns.boxplot(x='tip', y='day', hue='smoker', data=tip, dodge = True, color = "#e63091") 
plt.show()
```

## Adding meanprops(i.e, sign for identification of mean or median)

```python
# importing libraries
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np

titanic = sns.load_dataset('titanic')
titanic
titanic.head(9)# by varying the number inside perenthesis, we can limit the dataset as per requirement. 
p1 = sns.boxplot(x= 'survived', y='age', showmeans= True, meanprops= {"marker":"*","markersize":"12",
                                                                      "markeredgecolor":"beige"},data= titanic)
#show lables
plt.xlabel("The ones who survived", weight="bold",size='10')
plt.ylabel("Age(years)",weight="bold",size='10')
plt.title("The episode of Titanic",weight="bold",size='12')
plt.show()
```
