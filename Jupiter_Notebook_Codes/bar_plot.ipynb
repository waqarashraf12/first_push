```python
import seaborn as sns
from numpy import median 
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")
titanic
sns.barplot(x="sex", y="alone",hue="who", data = titanic, order=['female','child'], color='red', estimator = median,
           palette = "Dark2")
plt.show()
```
